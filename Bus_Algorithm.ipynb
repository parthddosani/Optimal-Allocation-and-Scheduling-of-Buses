```
trips = [2,5,6,5,6,7,6,6,4,5,5,5,4,5,4,5,4,4,4,4,3,2]
passengers = [34,152,231,441,567,632,478,413,217,189,246,189,176,256,239,310,345,389,256,113,79,25]
```

```
bus_timetable = {
    "table" : {
        "time" : [],
        "tag" : []
    }
}
passenger_count = []
```

```
def Bus_Schedule(trips,bus_timetable,curr_hr) :
  i = 0
  duration = 0
  count = 0
  while(i < len(trips)):
    duration = 1 / trips[i]
    count = passengers[i]/trips[i]
    j = 0
    while (j < trips[i]) :
      bus_timetable["table"]["time"].append(curr_hr + j*duration)
      bus_timetable["table"]["tag"].append(-1)
      passenger_count.append(count)
      j = j+1
    i = i + 1
    curr_hr = (curr_hr+1) % 24
  return bus_timetable
```
